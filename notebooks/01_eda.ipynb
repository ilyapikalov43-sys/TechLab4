# Exploratory Data Analysis (EDA) — Потребление ресурсов в жилых кварталах
## Анализ данных о электроэнергии, воде и газе

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/london_households.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
# Статистика по данным
df.describe()
```

```python
# Временной ряд потребления электроэнергии
plt.figure(figsize=(12, 5))
for household in df['household_id'].unique()[:3]:
    subset = df[df['household_id'] == household]
    plt.plot(subset['timestamp'], subset['energy_kwh'], label=f'Household {household}')
plt.xlabel('Time')
plt.ylabel('Energy (kWh)')
plt.title('Потребление электроэнергии по времени')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Корреляционная матрица
numeric_cols = ['energy_kwh', 'water_liters', 'gas_kwh', 'temperature_c', 'humidity_pct']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Корреляция между признаками')
plt.show()
```
